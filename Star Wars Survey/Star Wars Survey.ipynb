# Data Cleaning

## Removing NaN entries

```python
star_wars['RespondentID'].value_counts(dropna=False)
```

```python
star_wars = star_wars.dropna(subset=['RespondentID'])
star_wars['RespondentID'].value_counts(dropna=False)
```

## Creating boolean columns

```python
star_wars[['Have you seen any of the 6 films in the Star Wars franchise?',
           'Do you consider yourself to be a fan of the Star Wars film franchise?'
          ]]
```

```python
yes_no = {
    "Yes": True,
    "No": False
}
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'] = star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].map(yes_no)
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] = star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].map(yes_no)
```

```python
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].value_counts(dropna=False)
```

```python
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].value_counts(dropna=False)
```

```python
column_names = star_wars.columns[3:9]
column_names
```

For each of these columns, if the value in a cell is the name of the movie, that means the respondent saw the movie. If the value is NaN, the respondent either didn't answer or didn't see the movie. We'll assume that they didn't see the movie.

```python
for column in column_names:
    print(star_wars[column].value_counts(dropna=False))
```

```python
def convert_content_into_bool(val):
    return val in ["Star Wars: Episode I  The Phantom Menace",
        "Star Wars: Episode II  Attack of the Clones",
        "Star Wars: Episode III  Revenge of the Sith",
        "Star Wars: Episode IV  A New Hope",
        "Star Wars: Episode V The Empire Strikes Back",
        "Star Wars: Episode VI Return of the Jedi"]
```

```python
star_wars[star_wars.columns[3:9]] = star_wars[star_wars.columns[3:9]].applymap(convert_content_into_bool)
star_wars[star_wars.columns[3:9]].head(10)
```

Now we will convert the names of the columns in the list of **column_names** into more intuitive ones.

```python
i = 1
for column in column_names:
    star_wars.rename(columns={column: "seen_" + str(i)}, inplace=True)
    i += 1
star_wars[star_wars.columns[3:9]].head(10)
```

## Creating ranking columns

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
```

```python
star_wars[star_wars.columns[9:15]]
```

```python
i = 1
for column in star_wars.columns[9:15]:
    star_wars.rename(columns={column: "ranking_" + str(i)}, inplace=True)
    i += 1
star_wars[star_wars.columns[9:15]].head(10)
```

# Data Analysis

## Star Wars Movies Ranked By Respondants

First of all using the *mean()* method we will compute the mean for each ranking column. With this we can find the highest ranked movie quickly. The ranking goes from 1 to 5 being 1 the best.

```python
ranking = star_wars[star_wars.columns[9:15]].astype(float).mean(axis=0)
ranking
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(20, 10))
ax = star_wars[star_wars.columns[9:15]].astype(float).mean(axis=0).plot.bar(color=(0.2, 0.4, 0.6, 0.6))
ax.set_title("Star Wars Movies Ranked By Respondants", fontweight="bold", fontsize=15,)
ax.set_xlabel("Movie Rank",fontsize=12)
ax.set_ylabel("Ranking (1-5)",fontsize=12)
ax.tick_params(bottom=True, top=False, left=False, right=False, labelbottom=True)
```

In general, fans seem to prefer the original trilogy of the cinematic saga of Star Wars which consists of A New Hope (1977), The Empire Strikes Back (1980) and Return of the Jedi (1983). The trilogy was the first epic science fiction space adventure to be told on a large scale that is why it holds a special place in the hearts of Star Wars fans. The ranking_5 which corresponds to **Star Wars: Episode V The Empire Strikes Back** is considered by the respondents to be the best film in the saga, as we can see it is teh movie with the lowest mean. 

## Most Watched Star Wars Movie

```python
plt.figure(figsize=(20, 10))
ax = star_wars[star_wars.columns[3:9]].astype(float).sum(axis=0).plot.bar(color=(0.2, 0.4, 0.6, 0.6))
ax.set_title("Most Watched Star Wars Movie", fontweight="bold", fontsize=15,)
ax.set_xlabel("Movie Nmuber",fontsize=12)
ax.set_ylabel("Total Views",fontsize=12)
ax.tick_params(bottom=True, top=False, left=False, right=False, labelbottom=True)
```

As we have seen previously, **Star Wars: Episode V The Empire Strikes Back** is the Star Wars movie with the best rating. This fact is directly correlated with the number of views, as it is show in the **Most Watched Star Wars Movie** chart the fifth movie is the one with the most viewers.

## Respondant Differences By Gender 

```python
star_wars.groupby(by=['Gender'])[star_wars.columns[9:15]].mean()
```

```python
sns.set_style("darkgrid")
plt.rcParams["figure.figsize"] = [20, 10]
ax = star_wars.groupby(by=['Gender'])[star_wars.columns[9:15]].mean().transpose().plot.bar(
    color=[(0.2, 0.4, 0.6, 0.6), 'lightcoral'])
ax.set_title("Star Wars Movies Ranking By Gender", fontweight="bold", fontsize=15,)
ax.set_xlabel("Movie Rank",fontsize=12)
ax.set_ylabel("Ranking (1-5)",fontsize=12)
ax.tick_params(bottom=True, top=False, left=False, right=False, labelbottom=True)
```
