# Data Analysis

## Star Wars Movies Ranked By Respondants

First of all using the *mean()* method we will compute the mean for each ranking column. With this we can find the highest ranked movie quickly. The ranking goes from 1 to 5 being 1 the best.

```python
ranking = star_wars[star_wars.columns[9:15]].astype(float).mean(axis=0)
ranking
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(20, 10))
ax = star_wars[star_wars.columns[9:15]].astype(float).mean(axis=0).plot.bar(color=(0.2, 0.4, 0.6, 0.6))
ax.set_title("Star Wars Movies Ranked By Respondants", fontweight="bold", fontsize=15,)
ax.set_xlabel("Movie Rank",fontsize=12)
ax.set_ylabel("Ranking (1-5)",fontsize=12)
ax.tick_params(bottom=True, top=False, left=False, right=False, labelbottom=True)
```

In general, fans seem to prefer the original trilogy of the cinematic saga of Star Wars which consists of A New Hope (1977), The Empire Strikes Back (1980) and Return of the Jedi (1983). The trilogy was the first epic science fiction space adventure to be told on a large scale that is why it holds a special place in the hearts of Star Wars fans. The ranking_5 which corresponds to **Star Wars: Episode V The Empire Strikes Back** is considered by the respondents to be the best film in the saga, as we can see it is teh movie with the lowest mean.

# Star Wars Favorite Character

```python
star_wars[star_wars.columns[15:29]]
```

```python
star_wars['Han Solo'].value_counts(dropna=False)
```

First of all let's fill **NaN** value with the Unfamiliar (N/A) variable.

```python
star_wars[star_wars.columns[15:29]] = star_wars[star_wars.columns[15:29]].fillna('Unfamiliar (N/A)')
star_wars['Han Solo'].value_counts(dropna=False)
```

```python
def update_opinion(x):
    if x == "Very favorably":
        return 20
    elif x == "Somewhat favorably":
        return 10
    elif x == "Unfamiliar (N/A)" or x == "Neither favorably nor unfavorably (neutral)":
        return 0
    elif x == "Somewhat unfavorably":
        return -10
    elif x == "Very unfavorably":
        return -20
```

```python
char_scores = star_wars[star_wars.columns[15:29]]
char_scores = char_scores.applymap(update_opinion)
char_scores.head(5)
```

```python
plt.figure(figsize=(20, 10))
ax = char_scores.astype(float).sum(axis=0).plot.barh(color=(0.2, 0.4, 0.6, 0.6))
ax.set_title("Characters Likelihood", fontweight="bold", fontsize=15,)
ax.set_xlabel("Score",fontsize=12)
ax.set_ylabel("Character",fontsize=12)
ax.tick_params(bottom=True, top=False, left=False, right=False, labelbottom=True)
```

As we can see in the previous bar plot, **Han Solo** is the most liked character, followed by **Yoda** and **Obi Wan Kenobi**. On the other side the most unliked character is **Jar Jar Binks** followed by **Emperor Palpatine** and **Boba Fett**.

# Star Wars Most Controversial Character

To find the most controversial character with the following function we will split the answers:

```python
def split_rating_in_likes_and_dilikes(x):
    if x < 0:
        return 'Dislike'
    elif x == 0:
        return 'Neutral'
    else:
        return 'Like'
```

```python
char_scores = char_scores.applymap(split_rating_in_likes_and_dilikes)
char_scores.head(5)
```

```python
char_scores = char_scores.melt()
char_scores = char_scores[char_scores['value'] != 'Neutral']
char_scores.head(5)
```

```python
sns.set_style("darkgrid")
plt.rcParams["figure.figsize"] = [20, 10]
ax = char_scores.groupby(by=['variable', 'value']).size().transpose().plot.bar(
    color=[(0.2, 0.4, 0.6, 0.6), 'lightcoral'])
ax.set_title("Most Controversial Character", fontweight="bold", fontsize=15,)
ax.set_xlabel("Character",fontsize=12)
ax.set_ylabel("Like / Dislike Counts",fontsize=12)
ax.tick_params(bottom=True, top=False, left=False, right=False, labelbottom=True)
```

From the previous chart we can conclude that **Jar Jar Binks** is the most controversial chracter after comparing the likes and dislikes counts and it's followed by **Empeor Palpatine**.
